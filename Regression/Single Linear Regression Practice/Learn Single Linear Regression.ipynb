# Simple Linear Regression

# Import Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
df = pd.read_csv('/content/gdrive/My Drive/Salinan Salary_Data.csv')
df
```

# X is independent variable (YearsExperience)
# Y is independent variable (Salary) that going to predicted

```python
x = df.iloc[:,0]
y = df.iloc[:,1]
```

```python
x
```

```python
y
```

# Split test and train data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

# Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
salary_pred = lr.predict(x_test)
```

```python
lr.coef_
```

```python
plt.scatter(x_train,y_train,color='blue')
plt.plot(x_train,lr.predict(x_train),color='red')
plt.xlabel('YearsExperience')
plt.ylabel('Salary Prediction')
plt.show()
```

```python
plt.scatter(x_test,y_test,color='blue')
plt.plot(x_train,lr.predict(x_train),color='red')
plt.xlabel('YearsExperience')
plt.ylabel('Salary Prediction')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print(mean_squared_error(y_test,salary_pred))
print(r2_score(y_test,salary_pred))
```
