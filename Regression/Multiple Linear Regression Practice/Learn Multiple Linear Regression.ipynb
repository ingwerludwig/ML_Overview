# Import Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
df = pd.read_csv('/content/gdrive/My Drive/Salinan 50_Startups.csv')
df
```

# Split dependent and independent Variable

```python
x = df.iloc[:, 0:4].values
y = df.iloc[:, -1].values
```

```python
x
```

```python
y
```

# Encoding categorical Data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(categories='auto')
ct = ColumnTransformer(transformers=([('state_encoder',enc,[3])]),remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```python
print(x)
```

# Splitting Train and Test Dataset

```python
from sklearn.model_selection import train_test_split
X_test,X_train,Y_test,Y_train = train_test_split(x,y,train_size=0.2)
```

```python
Y_train
```

```python
Y_test
```

# Training Dataset

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,Y_train)
```

# Predict Test Dataset

```python
y_pred = lr.predict(X_test)
```

# Visualization Dependent Variable Test and Prediction

```python
def plotGraph(y_test,y_pred,regressorName):
    if max(y_test) >= max(y_pred):
        my_range = int(max(y_test))
    else:
        my_range = int(max(y_pred))
    plt.scatter(range(len(y_test)), y_test, color='blue')
    plt.scatter(range(len(y_pred)), y_pred, color='red')
    plt.title(regressorName)
    plt.show()
    return

plotGraph(Y_test, y_pred, "test")
```

# R2 Score Accuracy used to measure performance of continuous model

```python
from sklearn.metrics import r2_score
r2_score(Y_test,y_pred)
```
